```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np
# CHARACTERS
chars = np.load("cluster2/data.npz")
# print('The list of tokens:')
# print( chars['tokens'] )


# TSNE
model = TSNE()
plt.figure(figsize=(20,40))
embs = model.fit_transform(chars['points'])
plt.scatter(embs[:,0],embs[:,1],color='white')
for k in range(683): 
    plt.text(embs[k,0], embs[k,1], chars['tokens'][k])
plt.title('TSNE',size=18)
plt.savefig('tsne.png')
```
